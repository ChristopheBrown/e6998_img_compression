```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ssim_n_1 = pd.read_csv("ssim_n_1/history/history_.csv")
ssim_n_5 = pd.read_csv("ssim_n_5/history/history_.csv")
ssim_n_9 = pd.read_csv("ssim_n_9/history/history_.csv")
ssim_n_15 = pd.read_csv("ssim_n_15/history/history_.csv")

plt.title('Effect of n on Residual Loss')
plt.xlabel('epoch')
plt.ylabel('Residual Loss')
# plt.plot(ssim_n_1['residual_loss'], label='n=1')
plt.plot(ssim_n_5['residual_loss'], label='n=5')
plt.plot(ssim_n_9['residual_loss'], label='n=9')
plt.plot(ssim_n_15['residual_loss'], label='n=15')
plt.legend()
plt.savefig('iterations.png')
```

```python
ssim_batch_32 = 1 - pd.read_csv("ssim_batch_32/history/history_.csv")
ssim_batch_64 = 1 - pd.read_csv("ssim_batch_64/history/history_.csv")
ssim_batch_128 = 1 - pd.read_csv("ssim_batch_128/history/history_.csv")
ssim_batch_512 = 1 - pd.read_csv("ssim_batch_512/history/history_.csv")
ssim_batch_1024 = 1 - pd.read_csv("ssim_batch_1024/history/history_.csv")

plt.xlabel('epoch')
plt.ylabel('Accuracy')
plt.title('Training Accuacy (MS-SSIM) vs Epoch')
plt.plot(ssim_batch_32['SSIMLoss'], label='batch=32')
plt.plot(ssim_batch_64['SSIMLoss'], label='batch=64')
plt.plot(ssim_batch_128['SSIMLoss'], label='batch=128')
plt.plot(ssim_batch_512['SSIMLoss'], label='batch=512')
plt.plot(ssim_batch_1024['SSIMLoss'], label='batch=1024')
plt.legend()
plt.savefig('train_ssim_batch.png')
```
